###  Classify Movies Based on Ratings [Excellent, Good and Average]

```python
def rating(rating):
    if rating >=8:
        return 'Excellent'
    elif rating>=6:
        return 'Good'
    else:
        return 'Average'
```

```python
data['Classify Rates']=data['Rating'].apply(rating)
```

###  Find Unique Values From Genre 

```python
data['Genre'].nunique()
```

```python
list1=[]
for value in data['Genre']:
    list1.append(value.split(','))
```

```python
list1
```

```python
one_d=[]
for item in list1:
    for item1 in item:
        one_d.append(item1)
    
```

```python
uni_list=[]
for item in one_d:
    if item not in uni_list:
        uni_list.append(item)
        
```
